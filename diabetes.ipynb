```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
X = df.drop('diabetes', axis=1)
y = df['diabetes']
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_res,y_res,train_size=0.8,random_state=42,stratify=y_res)
```

```python
model=RandomForestClassifier(n_estimators=300,max_depth=20,min_samples_split=4,min_samples_leaf=2,class_weight='balanced',random_state=42)
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pickle

with open("diabetes_model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
from sklearn.metrics import f1_score
import pickle
import os

# Predict on test set
y_pred = model.predict(x_test)

# Calculate F1 score
f1 = f1_score(y_test, y_pred, average='macro')  # or 'weighted' if you prefer
print(f"Diabetes Model F1 Score: {f1:.2f}")

# Make sure metrics folder exists
os.makedirs("metrics", exist_ok=True)

# Save F1 score to diabetes_f1.txt
with open("metrics/diabetes_f1.txt", "w") as f:
    f.write(str(round(f1, 2)))

```
