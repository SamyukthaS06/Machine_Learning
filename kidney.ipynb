```python
from sklearn.preprocessing import StandardScaler
```

```python
num_cols = [
    'Blood pressure (mm/Hg)', 'Random blood glucose level (mg/dl)',
    'Blood urea (mg/dl)', 'Serum creatinine (mg/dl)', 'Sodium level (mEq/L)',
    'Potassium level (mEq/L)', 'Hemoglobin level (gms)',
    'Red blood cell count (millions/cumm)', 'Estimated Glomerular Filtration Rate (eGFR)',
    'Urine protein-to-creatinine ratio', 'Cystatin C level'
]

scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['Target'],axis=1)
y=df['Target']
```

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)
```

```python
x_train,x_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.2, random_state=42, stratify=y_res)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(
    objective='multi:softmax',
    eval_metric='mlogloss',
    use_label_encoder=False,
    random_state=42,
    n_estimators=300,
    max_depth=7,
    learning_rate=0.1
)
```

```python
from sklearn.metrics import classification_report
```

```python
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
import pickle
with open("kidney_model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
from sklearn.metrics import f1_score
import pickle
import os

# Example: You already split your data like this
# X_train, X_test, y_train, y_test

# Prediction
y_pred = model.predict(x_test)

# F1 Score
f1 = f1_score(y_test, y_pred, average='macro')  # Use 'weighted' if needed
print(f"F1 Score: {f1:.2f}")

# Create 'metrics' folder if it doesn't exist
os.makedirs("metrics", exist_ok=True)

# Save F1 score to a file
with open("metrics/kidney_f1.txt", "w") as f:
    f.write(str(round(f1, 2)))

```
